```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
patient_data = pd.read_csv("/content/patients-patients.csv")
patient_data
```

# New Section

**Data Cleaning**

```python
patient_data.drop(['id', 'country', 'department', 'status'], axis=1, inplace=True)
patient_data.fillna(0)
```

```python
#To remove duplicates in data
patient_data.drop_duplicates(inplace = True)
```

```python
patient_data.isnull().any()
```

```python
patient_data.birth_date.fillna(patient_data.median()).isnull().any()
patient_data.birth_date.fillna(patient_data.mean()).isnull().any()
```
